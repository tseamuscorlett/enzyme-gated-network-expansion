```python
from Bio.PDB import *
```

```python
HUP_pdbs = []
with open('../HUP_pdbs.txt' ,'r') as file:
    lines = file.readlines()
    for line in lines:
        HUP_pdbs.append(line.strip())
```

```python
HUP_pdbs
```

```python
for pdb_id in HUP_pdbs:
# download PDB structure file
    pdb_file_path = PDBList().retrieve_pdb_file(pdb_id, file_format='pdb', pdir='.')
    
# get structure
    structure = PDBParser().get_structure(pdb_id, pdb_file_path)
    
# get info from structure
    resolution = structure.header['resolution']
    length = len(list(structure.get_residues()))

    print(f"PDB ID: {pdb_id}")
    print(f"Resolution: {resolution} Å")
    print(f"Length: {length} residues")
```

```python
import requests

# Define the list of example PDB IDs
pdb_ids = HUP_pdbs[:3]

# Iterate through the PDB IDs and retrieve information
for pdb_id in pdb_ids:
    url = f'https://data.rcsb.org/rest/v1/core/entry/{pdb_id}'

    # Make a GET request to the PDB API
    response = requests.get(url)

    # Check if the request was successful (status code 200)
    if response.status_code == 200:
        # Extract resolution and sequence length from the response JSON
        data = response.json()
        resolution = data['rcsb_entry_info']['resolution_combined']
        sequence_length = data['rcsb_entry_info']['deposited_polymer_monomer_count']

        # Print the information
        print(f'PDB ID: {pdb_id}')
        print(f'Resolution: {resolution}')
        print(f'Sequence Length: {sequence_length}\n')
    else:
        print(f'Failed to retrieve data for PDB ID {pdb_id}')
```
