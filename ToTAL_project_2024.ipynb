```python
import re
import numpy as np
import matplotlib.pyplot as plt
from matplotlib_venn import venn2

def clean_text(text):
    """
    Returns a new string where non-letters are removed.
    Also converts capital letters to lower-case.
    """
    
    cleaned_text = re.sub(r'[^a-zA-Z\s—-]', '', text)
    return cleaned_text.lower()


def get_word2count(source):
    common_words = ['the', 'of', 'to', 'and', 'a', 'that', 'in', 'is', 'be', 'his', 'or', 'he', 'as', 'it', 'not', 'for', 'may', 'are', 'will', 'an', 'who', 'their', 'one', 'can', 'by', 'but', 'with', 'they', 'have', 'we', 'on', 'must', 'if', 'at', 'him', 'from', 'them', 'no', 'some', 'do', 'such', 'i', 'our', 'need', 'seem', 'people', 'has', 'all', 'you', 'so', 'what', 'make', 'which', 'how', 'these', 'out', 'other', 'should', 'see', 'more', 'this', 'each', 'often', 'every', 'never', 'even', 'others', 'get', 'only', 'would', 'another', 'take', 'been', 'many', 'much', 'into', 'when', 'than', 'my', 'into', 'any', 'very', 'those', 'might', 'too', 'being', 'still', 'within', 'does', 'made'
'without', 'lot', 'like', 'think', 'its', 'dont', 'was', 'just', 'because', 'youre', 'know', 'then', 'say', 'good', 'about', 'things', 'your', 'theyre', 'me', 'were', 'doing', 'feel', 'im', 'there', 'had', 'really', 'actually', 'also', 'getting', 'something', 'okay', 'yeah', 'theres', 'uh', 'thing', 'already', 'maybe', 'um', 'without', 'thats', 'hes', 'sometimes', 'while', 'oh', '—', 'yes', 'towards', '', 'ive', 'youll', 'mh', 'could', 'during', 'made', 'guess', 'sort', 'whats', 'around', 'over', 'here', 'one', 'two', 'three', 'id', 'instead', 'did', 'didnt', 'however']
    word2count = {}
    
    with open(source, 'r') as file:
        lines = file.readlines()
        for line in lines:
            words = line.strip().split()
            for word in words:
                clean_word = clean_text(word)
                if clean_word not in common_words:
                    if clean_word not in word2count:
                        word2count[clean_word] = 1
                    else:
                        word2count[clean_word] += 1
    
    sorted_dict = dict(sorted(word2count.items(), key=lambda item: item[1], reverse=True))
    
    return sorted_dict
```

# chi-square

```python
all_combo = {}
for k,v in academic_combo.items():
    all_combo[k] = [v, 0]

for k,v in business_combo.items():
    if k not in all_combo:
        all_combo[k] = [0, v]
    else:
        all_combo[k] = [academic_combo[k], v]
```

```python
import scipy.stats as stats

# Observed frequencies
observed = list(all_combo.values())

# Perform Chi-Square test
chi2, p, dof, ex = stats.chi2_contingency(observed)

print(f"Chi-Square: {chi2}")
print(f"P-Value: {p}")
```

```python
def histogram(dict1, val_type = 'MEAN', bins = 10, x_axis = 'x-axis', y_axis ='counts'):
    data1 = list(dict1.values())
    
    if type(data1[0]) == dict:
        data1 = [x[val_type] for x in data1]
    
    plt.hist(data1, bins=bins, edgecolor='k')
    plt.xlabel(x_axis)
    plt.ylabel(y_axis)
    plt.show()
```
