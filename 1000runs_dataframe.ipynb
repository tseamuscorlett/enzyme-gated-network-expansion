```python
import glob
import re
import pandas as pd

def makeDF(list_of_dicts, runDates, filename):
    # concatenate dicts into a single df
    transDFs = []
    for d in list_of_dicts:
        transDFs.append(pd.DataFrame([d]).T)  # transpose

    optimal_df = pd.concat(transDFs, axis=1)
    optimal_df.columns = runDates
    optimal_df.to_csv(filename)
```

```python
makeDF(dicts1, runSEEDs, 'cpds_folditer_preATP.csv')
makeDF(dicts2, runSEEDs, 'folds_folditer_preATP.csv')
makeDF(dicts3, runSEEDs, 'rns_folditer_preATP.csv')
makeDF(dicts4, runSEEDs, 'rules_folditer_preATP.csv')

makeDF(dicts5, runSEEDs, 'cpds_cumiter_preATP.csv')
makeDF(dicts6, runSEEDs, 'folds_cumiter_preATP.csv')
makeDF(dicts7, runSEEDs, 'rns_cumiter_preATP.csv')
makeDF(dicts8, runSEEDs, 'rules_cumiter_preATP.csv')
```
