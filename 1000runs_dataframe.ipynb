```python
import glob
import re
import pandas as pd

def makeDF(list_of_dicts, runDates, filename):
    # concatenate dicts into a single df
    transDFs = []
    for d in list_of_dicts:
        transDFs.append(pd.DataFrame([d]).T)  # transpose

    optimal_df = pd.concat(transDFs, axis=1)
    optimal_df.columns = runDates
    optimal_df.to_csv(filename)
```

# random order

```python
folder_path = 'fold_results/random_order/'
files_pattern = '*.pkl.gz'
files = glob.glob(folder_path + files_pattern)

dicts1 = []  # list of dicts
dicts2 = []
dicts3 = []
dicts4 = []
dicts5 = []
dicts6 = []
dicts7 = []
dicts8 = []

runSEEDs = []
count = 1

# Iterate over each file
for file_path in files:
    
    # Extract the RANDOMSEED, append to list
    runSEEDs.append(file_path.split('_')[-1].split('.')[0])
    
    print(count, file_path.split('_')[-1].split('.')[0])
    count += 1

    # Read the pickle file, append to list
    result = pd.read_pickle(file_path)
    dicts1.append(result.cpds_folditer)  # dict {cpd: folditer}
    dicts2.append(result.folds_folditer)  # dict {fold: folditer}
    dicts3.append(result.rns_folditer)  # dict {rn: folditer}
    dicts4.append(result.rules_folditer)  # dict {rule: folditer} 
    
    dicts5.append(result.cpds_cumiter)  # dict {cpd: cumiter}
    dicts6.append(result.folds_cumiter)  # dict {fold: cumiter}
    dicts7.append(result.rns_cumiter)  # dict {rn: cumiter}
    dicts8.append(result.rules_cumiter) # dict {rule: cumiter}

# export csv files
makeDF(dicts1, runSEEDs, 'cpds_folditer_random_order.csv')
makeDF(dicts2, runSEEDs, 'folds_folditer_random_order.csv')
makeDF(dicts3, runSEEDs, 'rns_folditer_random_order.csv')
makeDF(dicts4, runSEEDs, 'rules_folditer_random_order.csv')
makeDF(dicts5, runSEEDs, 'cpds_cumiter_random_order.csv')
makeDF(dicts6, runSEEDs, 'folds_cumiter_random_order.csv')
makeDF(dicts7, runSEEDs, 'rns_cumiter_random_order.csv')
makeDF(dicts8, runSEEDs, 'rules_cumiter_random_order.csv')
```
