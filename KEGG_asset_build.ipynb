```python
import csv
import ast
import requests
```

```python
def dict2csv(my_dict, csv_file_path):
    with open(csv_file_path, 'w', newline='') as csv_file:
        csv_writer = csv.writer(csv_file)
        for key, value in my_dict.items():
            csv_writer.writerow([key] + [value])

def csv2dict(csv_file_path):
    result_dict = {}
    with open(csv_file_path, 'r') as csv_file:
        csv_reader = csv.reader(csv_file)
        
        for row in csv_reader:
            try:
                result_dict[row[0]] = ast.literal_eval(row[1])  # value is list
            except:
                result_dict[row[0]] = row[1]
                
    return result_dict
```

# cpd2reac, cpd2prod, cpd2rn

```python
cpd2reac = {}
cpd2prod = {}

for c in cpd2nameShort.keys():
    cpd2reac[c] = []
    cpd2prod[c] = []

    for rn, eqn in rn2eqn_SI.items():
        if c in eqn.split(' => ')[0]:  # compound is a reactant
            cpd2reac[c].append(rn)
        if c in eqn.split(' => ')[1]:  # compound is a product
            cpd2prod[c].append(rn)
```

```python
cpd2rn = {}

for c in cpd2nameShort.keys():
    cpd2rn[c] = []

    for rn, eqn in rn2eqn_SI.items():
        if c in eqn.split(' => ')[0] or c in eqn.split(' => ')[1]:  # compound is a reactant or product
            cpd2rn[c].append(rn)
```

# rn2ec

```python
rn2ec = {}
file_path = '../data/assets/KEGG_ec2rn.txt'

with open(file_path, 'r') as file:
    lines = file.readlines()
    
    for line in lines:
        ec = line.split('\t')[0][3:]
        rn = line.split('\t')[1].strip()[3:]
    
        rn2ec[rn] = ec  # some reactions appear more than once; just overwrite
```
