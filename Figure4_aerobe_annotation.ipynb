**NOTE:** Some variable names differ from those used in the main text, 
reflecting terminology changes made during the course of the study:
- 'fold' -> 'lineage'
- 'NONE' -> 'base seed'
- 'ALL' -> 'full network'

Each enzyme-gated network expansion run records iterations of network expansion as follows:
- cumiter: Tracks all individual expansion steps, including both lineage addition steps and the resulting standard network expansion steps
- folditer: Groups each lineage addition and the subsequent network expansion into a single unit; corresponds to a metabolic stage

# LTU

```python
aerobe2hits_LTU = {}
anaerobe2hits_LTU = {}
facultative2hits_LTU = {}

for g in checkA:
    categoryHits = 0
    allHits = 0
    for x in genome2xfull[g]:
        if x in LTU:
            categoryHits += genome2xfull[g][x]
        allHits += genome2xfull[g][x]
    aerobe2hits_LTU[g] = categoryHits / allHits

for g in checkF:
    categoryHits = 0
    allHits = 0
    for x in genome2xfull[g]:
        if x in LTU:
            categoryHits += genome2xfull[g][x]
        allHits += genome2xfull[g][x]
    facultative2hits_LTU[g] = categoryHits / allHits

for g in checkN:
    categoryHits = 0
    allHits = 0
    for x in genome2xfull[g]:
        if x in LTU:
            categoryHits += genome2xfull[g][x]
        allHits += genome2xfull[g][x]
    anaerobe2hits_LTU[g] = categoryHits / allHits
    
print(len(aerobe2hits_LTU))

# Determine the range for the bins that covers 3 datasets
min_val = min(min(aerobe2hits_LTU.values()), min(anaerobe2hits_LTU.values()), min(facultative2hits_LTU.values()))
max_val = max(max(aerobe2hits_LTU.values()), max(anaerobe2hits_LTU.values()), max(facultative2hits_LTU.values()))
bins = np.linspace(min_val, max_val, 100)

plt.hist(aerobe2hits_LTU.values(), bins=bins, color='r', label='aerobe', alpha=0.5)  
plt.hist(facultative2hits_LTU.values(), bins=bins, color='black', label='facultative', alpha=1.0)  
plt.hist(anaerobe2hits_LTU.values(), bins=bins, color='blue', label='anaerobe', alpha=0.5) 

# U-test
statistic, p_value = mannwhitneyu(list(aerobe2hits_LTU.values()), list(anaerobe2hits_LTU.values()))
p_value = '%.2E' % Decimal(p_value)
plt.annotate(f'p = {p_value}', xy=(0.85, 0.55), xycoords='axes fraction', ha='center', va='center', fontsize=10, color='purple')


plt.xlabel('LTU fold hits / all hits')
plt.ylabel('# genomes')
plt.legend()
# plt.savefig('histograms_LTU.svg', dpi=300, bbox_inches = 'tight')
plt.show()
```
