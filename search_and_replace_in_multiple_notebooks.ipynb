```python
import os
import nbformat

def search_and_replace_in_notebooks(directory, search_text, replace_text):
    for filename in os.listdir(directory):
        if filename.endswith('.ipynb') and not filename.startswith('search_and_replace'):
            file_path = os.path.join(directory, filename)
            print(f'Processing file: {file_path}')
            
            # Load the notebook
            with open(file_path, 'r', encoding='utf-8') as f:
                nb = nbformat.read(f, as_version=4)

            # Search and replace in each cell
            modified = False
            for cell in nb.cells:
                if cell.cell_type == 'code':
                    if search_text in cell.source:
                        cell.source = cell.source.replace(search_text, replace_text)
                        modified = True
                elif cell.cell_type == 'markdown':
                    if search_text in cell.source:
                        cell.source = cell.source.replace(search_text, replace_text)
                        modified = True

            # Save the modified notebook if changes were made
            if modified:
                with open(file_path, 'w', encoding='utf-8') as f:
                    nbformat.write(nb, f)
                print(f'Updated file: {file_path}')

# Example usage
# directory = '../networkExpansion/'
# search_text = 'replaceThisBaby'
# replace_text = 'replaced!'
# search_and_replace_in_notebooks(directory, search_text, replace_text)
```

```python
# /data/assets/xgroup2/
directory = '../networkExpansion/'
search_text = '../assets'
replace_text = '../data/assets/'
search_and_replace_in_notebooks(directory, search_text, replace_text)
```
